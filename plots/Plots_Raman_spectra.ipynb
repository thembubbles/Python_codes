# Raman spectra plotting


This script reads text files containing Raman spectroscopy measurements and creates spectra plots using the matplotlib package

**Script written by Fernando Araujo (KU Leuven)**

```python
# ---------------- plotting Raman data from text files
        
# --- First step is to import the modules that will be used in the script

import os
import glob  #These two modules are used to look into folders in the PC and access the files


import numpy as np # This module is used for data manipulation


import matplotlib.pyplot as plt # This module is what we will use for plotting the spectra

#You can check the documentation in the link below to know more about the parameters I am using in the plot
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html




# --- set working directory: Choose the folder that contains your spectra text files

data_dir = './_DATA/Marijke'
```

## Raman data

```python
# --- look into working directory and create a list with selected files - .txt

os.chdir(data_dir) 
# This line uses the 'os' package that we imported before 
#to look into the folder that contains the spectra files

Raman_files = glob.glob('*.txt')
# This line uses the 'glob' package to create a list containing the name of all the spectra files



# You can also create specific lists separating the files according to their orientation or source

Ethiopia_files = glob.glob('e*.txt') #use files starting with an 'e'
Russia_files = glob.glob('r*.txt') #use files starting with an 'r'
Zambia_files = glob.glob('z*.txt') #use files starting with a 'z'


C_orientation_files = glob.glob('*_c_*.txt') #use files containing '_c_' in the name
O_orientation_files = glob.glob('*_0_*.txt') #use files containing '_0_' in the name



print(Ethiopia_files) #This line shows the list with spectra file names in the output 
len(Ethiopia_files) #Shows the size of the list (i.e., number of spectra files)
```

```python
# --- In this cell we will get the data from the spectra file using the list we created and then plot it
# --- You only need to change the items in the first section (i.e., list to be used and files to plot)


    # -- 1) This section has the parts where you will need to change everytime you want to make a plot

#indicate which list should be used    
file_list_to_use = Ethiopia_files 


#indicate the name of the files you want to plot
file1 = 'E_100322690344_B_Orientation_0_1800 (450-850nm)_532nm_Edge_100%_x100_VIS_LWD_H60um_20sX3_01.txt'

file2 = 'E_100322690344_B_Orientation_45_1800 (450-850nm)_532nm_Edge_100%_x100_VIS_LWD_H60um_20sX3_01.txt'

file3 = 'e_100322690344_b_orientation_90_1800 (450-850nm)_532nm_edge_100%_x100_vis_lwd_h60um_20sx3_01.txt'

file4 = 'e_100322690344_b_orientation_c_1800 (450-850nm)_532nm_edge_100%_x100_vis_lwd_h60um_20sx3_01_0.txt'






    # -- 2) This section reads the text files and import the data, one line for each file selected above
    
spectrum1 = np.genfromtxt(file_list_to_use[file_list_to_use.index(file1)],comments='#', delimiter='\t')
spectrum2 = np.genfromtxt(file_list_to_use[file_list_to_use.index(file2)],comments='#', delimiter='\t')
spectrum3 = np.genfromtxt(file_list_to_use[file_list_to_use.index(file3)],comments='#', delimiter='\t')
spectrum4 = np.genfromtxt(file_list_to_use[file_list_to_use.index(file4)],comments='#', delimiter='\t')







    # -- 3) This section creates the plot with the imported spectra

plt.figure(figsize=(16,12)) #code for controling figure size (x,y)

#codes to plot each spectra
plt.plot(spectrum1[:,0],spectrum1[:,1]+300,color='green', linestyle='-', linewidth=4, label='Orientation 0')
plt.plot(spectrum2[:,0],spectrum2[:,1]+200,c='green',ls='--', lw=4, label='Orientation 45')
plt.plot(spectrum3[:,0],spectrum3[:,1]+100,c='green',ls='-.', lw=4, label='Orientation 90')
plt.plot(spectrum4[:,0],spectrum4[:,1],c='green',ls=':', lw=4, label='Orientation C')

# plt.plot() is the command used, everything insided the parenthesis is a parameter of the plot
# plt.plot(x values, y values, line color, line style, line thickness, line label)

# You can change the stacking by changing the value (+300, +200, +100, etc.) after the second parameter (y values)
# You can change the line color by typing the color name in the c (or color) parameter
# You can change the line style by typing the style in the ls (or linestyle) parameter
# You can change the line thickness by typing another value in the lw (or linewidth) parameter
# You can change the label in the legend by typing in the label parameter


#codes to change the title of the axes and of the plot itself
plt.xlabel("Raman shift, cm$^{-1}$", fontsize = 16)
plt.ylabel("intensity, arbitrary units", fontsize = 16)
plt.title("Spectra of ethiopian sample '100322690344_B' according to changing orientation",fontsize = 20,fontweight="bold")

#code to fix the range of the axes
plt.xlim(200,1500)
# plt.ylim(0,5000)

#code to change the fontsize in the ticks of the axes
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)

#code to change legend parameters
plt.legend(fontsize = 16, loc='best')

plt.show()
```
